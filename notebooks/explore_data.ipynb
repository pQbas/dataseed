```python
import pandas as pd
```

```python
df = pd.read_csv('/home/pqbas/projects/LABINM_Robotics_Automation/SeedlingsNet/classifiers/data/dataset/data.csv')
print(df.to_string()) 
```

```python
lengths = df.loc[:,"length"]
lengths_array = lengths.to_numpy()
print(lengths.to_string())
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

print("mean:",np.mean(lengths_array))
print("stdv:",np.std(lengths_array))

plt.hist(lengths_array, color='lightgreen', ec='black', bins=50)
plt.axvline(np.mean(lengths_array), color='red')
plt.xlabel('length of bigger leaf [mm]')
plt.title('Distribution of lengths of bigger leafs')
plt.plot(np.arange(0, 150, 0.01) , 500*norm.pdf(np.arange(0, 150, 0.01), np.mean(lengths_array), np.std(lengths_array)))
plt.show()
```
